```python
import numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import time
import tracemalloc
import linecache
import os
import seaborn as sns
```

```python
cell2loc_pcc = pd.read_csv("PCC/cell2loc_pcc.csv", index_col=0)
dwls_pcc = pd.read_csv("PCC/dwls_pcc.csv", index_col=0)
ground_pcc = pd.read_csv("PCC/ground_pcc.csv", index_col=0)
spotlight_pcc = pd.read_csv("PCC/spotlight_pcc.csv", index_col=0)

cell2loc_spearman = pd.read_csv("SPEARMANR/cell2loc_spearmanr.csv", index_col=0)
dwls_spearman = pd.read_csv("SPEARMANR/dwls_spearmanr.csv", index_col=0)
ground_spearman = pd.read_csv("SPEARMANR/ground_spearmanr.csv", index_col=0)
spotlight_spearman = pd.read_csv("SPEARMANR/spotlight_spearmanr.csv", index_col=0)
```

```python
cell2loc_pcc
```

```python
#Difference Matrix
diff_matrix = np.abs(cell2loc_pcc - cell2loc_spearman)


#Heatmaps
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.heatmap(cell2loc_pcc, annot=True)
plt.title('Pearson Correlation')
plt.subplot(1, 2, 2)
sns.heatmap(cell2loc_spearman, annot=True)
plt.title('Spearman Correlation')
plt.show()

#Difference Heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(diff_matrix, annot=True)
plt.title('Absolute Difference Matrix')
plt.show()
```

```python
#Difference Heatmap
plt.figure(figsize=(10, 10))
ax = sns.heatmap(diff_matrix, xticklabels=False)
# Rotate y-axis labels
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, ha="right")

plt.title('Cell2loc Absolute Difference Matrix')
plt.show()
```

```python
diff_matrix = np.abs(dwls_pcc - dwls_spearman)
#Difference Heatmap
plt.figure(figsize=(10, 10))
ax = sns.heatmap(diff_matrix, xticklabels=False)
# Rotate y-axis labels
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, ha="right")

plt.title('DWLS Absolute Difference Matrix')
plt.show()
```

```python
diff_matrix = np.abs(spotlight_pcc - spotlight_spearman)
#Difference Heatmap
plt.figure(figsize=(10, 10))
ax = sns.heatmap(diff_matrix, xticklabels=False)
# Rotate y-axis labels
ax.set_yticklabels(ax.get_yticklabels(), rotation=0, ha="right")

plt.title('Spotlight Absolute Difference Matrix')
plt.show()
```
