```python
def process_data(mat, allnames):
    
    n_slides = mat['tot'].shape[0]
    n_channels = mat['tot'].shape[1]

    tot_data = mat['tot']
    gene_names = allnames['allNames'].flatten()
    
    # Reshape the data
    reshaped_data = tot_data.reshape(len(gene_names), -1)

    # Create a DataFrame
    df = pd.DataFrame(reshaped_data, columns=[f'Slide_{s+1}_Channel_{c+1}' for s in range(n_slides) for c in range(n_channels)])
    df['Gene_Name'] = gene_names  # Add gene names as a new column

    # Reorder columns to have Gene_Name first
    cols = ['Gene_Name'] + [f'Slide_{s+1}_Channel_{c+1}' for s in range(n_slides) for c in range(n_channels)]
    df = df[cols]

    return df

def flatten_to_float(arr):
    if len(arr) > 0:
        flattened_array = np.array(arr).flatten()
        if len(flattened_array) > 0:
            return flattened_array[0]
    return np.nan  # Return NaN if the array is empty or cannot be flattened
```

```python
# Process the new data using the defined function
df2 = process_data(RNA_loc_2_mat, allgeneNames_mat)

# Apply the flatten_to_float function to the DataFrame
df2_mod = df2.map(flatten_to_float)

# Print the DataFrame with flattened values as floats
print(df2_mod.head())
```

```python
df2_mod.to_csv('RNA_rep_2_spatial_data.csv', index = False)
```
